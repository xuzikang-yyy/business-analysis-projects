```python
import pandas as pd
import matplotlib.pyplot as plt
import os

plt.rcParams['font.sans-serif'] = ['Arial Unicode MS']
plt.rcParams['axes.unicode_minus'] = False

# 模拟财务数据
data = {
    'product': ['A', 'B', 'C', 'D', 'E'],
    'revenue': [120000, 85000, 230000, 150000, 70000],
    'variable_cost': [50000, 35000, 90000, 65000, 30000],
    'fixed_cost': [30000, 20000, 45000, 35000, 15000]
}
df = pd.DataFrame(data)

# 财务指标计算
df['profit'] = df['revenue'] - df['variable_cost'] - df['fixed_cost']
df['margin'] = df['profit'] / df['revenue']
df['cm'] = df['revenue'] - df['variable_cost']

total_fc = df['fixed_cost'].sum()
total_cm = df['cm'].sum()
break_even = total_fc / (total_cm / df['revenue'].sum())

os.makedirs('02_企业财务成本与利润分析', exist_ok=True)

# 可视化
plt.figure(figsize=(10, 5))
plt.bar(df['product'], df['profit'], color='green')
plt.title('各产品线利润')
plt.ylabel('利润')
plt.tight_layout()
plt.savefig('02_企业财务成本与利润分析/profit.png')
plt.close()

# 输出结果
print('='*50)
print(f"盈亏平衡点收入：{break_even:.2f} 元")
print('='*50)
print(df)
```
