```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from datetime import datetime
import os

plt.rcParams['font.sans-serif'] = ['Arial Unicode MS']
plt.rcParams['axes.unicode_minus'] = False

# 模拟 30 天运营数据
date_range = pd.date_range(start='2024-03-01', periods=30, freq='D')
data = {
    'date': date_range,
    'uv': np.random.randint(800, 2000, 30),
    'pv': np.random.randint(3000, 8000, 30),
    'order': np.random.randint(50, 150, 30)
}
df = pd.DataFrame(data)

# 计算核心指标
df['ctr'] = df['order'] / df['uv']
daily_uv = df['uv'].mean()
daily_order = df['order'].sum()

# 自动创建文件夹（不报错）
os.makedirs('03_运营数据自动化报表', exist_ok=True)

# 画图
plt.figure(figsize=(12, 5))
plt.plot(df['date'], df['uv'], label='UV', marker='o')
plt.plot(df['date'], df['order'], label='订单量', marker='s')
plt.title('运营日报趋势')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('03_运营数据自动化报表/operation_trend.png')
plt.close()

# 生成自动化报表
report = f"""
运营自动化日报
生成时间：{datetime.now().strftime('%Y-%m-%d %H:%M')}
日均UV：{daily_uv:.0f}
总订单：{daily_order}
报表图表已保存
"""

# 保存到文件
with open('03_运营数据自动化报表/report.txt', 'w', encoding='utf-8') as f:
    f.write(report)

# 打印输出
print(report)
```
